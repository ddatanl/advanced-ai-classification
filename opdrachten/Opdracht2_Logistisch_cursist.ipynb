## 3. Beslissingsgrens visualiseren
Met twee features kunnen we het hele feature-vlak inkleuren op basis van de voorspelde klasse.
De code is al voor je geschreven — voer hem uit.

```python
# Beslissingsgrens-plot (al ingevuld)
# We maken een raster van punten in het feature-vlak en voorspellen voor elk punt de klasse.
X_all_2 = X[twee_features].values
x_min, x_max = X_all_2[:,0].min()-0.4, X_all_2[:,0].max()+0.4
y_min, y_max = X_all_2[:,1].min()-0.3, X_all_2[:,1].max()+0.3
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
Z = model_2.predict(np.c_[xx.ravel(), yy.ravel()])
# Map labels to integers for plotting
labels_unique = list(model_2.classes_)
Z_int = np.array([labels_unique.index(z) for z in Z]).reshape(xx.shape)

cmap_bg = ListedColormap(["#DCE6F8", "#FBE3CC", "#CFD4E0"])
fig, ax = plt.subplots(figsize=(8, 5))
ax.contourf(xx, yy, Z_int, alpha=0.5, cmap=cmap_bg)
palette = {"setosa": "#4473C5", "versicolor": "#D97733", "virginica": "#203864"}
for naam, kleur in palette.items():
    mask = (y == naam)
    ax.scatter(X.loc[mask, twee_features[0]], X.loc[mask, twee_features[1]],
               color=kleur, s=40, alpha=0.9, edgecolor="white", linewidth=0.5, label=naam)
ax.set_xlabel(twee_features[0]); ax.set_ylabel(twee_features[1])
ax.set_title("Beslissingsgrens (logistische regressie, 2 features)")
ax.legend(); plt.show()
```

> **Bekijk de plot:** je ziet drie gekleurde gebieden — één per klasse. De grenzen ertussen zijn
> rechte lijnen, een typische eigenschap van logistische regressie.
